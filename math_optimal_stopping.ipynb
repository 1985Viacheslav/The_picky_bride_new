# Стратегия:

Оптимальная стратегия для задачи выбора невесты основана на принципе "разведки и эксплуатации"

Разделим процесс на две фазы: фазу наблюдения и фазу выбора

В фазе наблюдения мы отвергаем определенное количество претендентов, запоминая лучшего среди них

В фазе выбора мы выбираем первого претендента, который лучше всех, кого мы видели в фазе наблюдения

Ключевой момент стратегии - определение оптимальной точки перехода между фазами

Математически доказано, что оптимальная точка перехода находится на расстоянии 1/e от начала последовательности, где e - число Эйлера (приблизительно 2.71828)

Эта стратегия гарантирует вероятность выбора лучшего жениха, стремящуюся к 1/e (около 37%) при увеличении числа претендентов

## Детали реализации:

Создадим список претендентов, где каждый претендент представлен уникальным числом (чем меньше число, тем лучше претендент)

Перемешаем список претендентов для имитации случайного порядка их появления

Вычислим точку перехода как n/e, где n - общее число претендентов

В фазе наблюдения (до точки перехода) будем отслеживать лучшего претендента

В фазе выбора будем сравнивать каждого претендента с лучшим из фазы наблюдения

Выберем первого претендента, который лучше всех из фазы наблюдения, и он станет женихом

Если такого не найдем до конца, выберем последнего претендента, который станет женихом

```python
import random
import math

def optimal_choice(n):
    # Создаем список претендентов, где 1 - лучший, n - худший
    applicants = list(range(1, n + 1))
    
    # Перемешиваем список для имитации случайного порядка появления претендентов
    random.shuffle(applicants)
    
    # Вычисляем точку перехода (n/e с округлением вниз до целого значения)
    cutoff = int(n / math.e)
    
    # Фаза наблюдения: находим лучшего среди первых cutoff претендентов
    best_tmp = min(applicants[:cutoff])
    
    # Фаза выбора: выбираем первого претендентa лучше, чем best_tmp
    for i in range(cutoff, n):
        if applicants[i] < best_tmp:
            return applicants[i] == 1  # Возвращаем True, если выбран лучший претендент (1)
    
    # Если не нашли никого лучше, выбираем последнего
    return applicants[-1] == 1  # Возвращаем True, если последний - лучший

def simulate(n, simulations):
    # Проводим множество симуляций и подсчитываем успешные исходы
    successes = sum(optimal_choice(n) for _ in range(simulations))
    
    # Возвращаем долю успешных исходов
    return successes / simulations

# Параметры для симуляции
n = 100  # Количество претендентов
simulations = 100000  # Количество симуляций для получения статистически значимого результата

# Запускаем симуляцию и выводим результат
success_rate = simulate(n, simulations)
print(f"Вероятность выбора лучшего претендента на роль жениха: {success_rate:.2%}")
```
